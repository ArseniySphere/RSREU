**Задание 5. Законы распределения дискретных случайных величин.** 

А - биномальный закон распределения.

Имеется шесть пробирок с питательной средой. Вероятность
появления колонии микроорганизмов определённого вида в какой-либо из
шести пробирок постоянна и равна р=0,7.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

def bin_prob(n, p):
    probs = []
    for m in range(n + 1):
        one_prob = comb(n, m) * p ** m * (1 - p) ** (n - m)
        probs.append(one_prob)
    return probs
    

n = 6
p = 0.7
probabilities = bin_prob(n, p)
x = list(range(n + 1))

print("Закон распределения:")
for i in range(n + 1):
    print(f"{i:<10} {probabilities[i]:<10.4}")

M = n * p
D = n * p * (1 - p)
b = np.sqrt(D)
mx = max(probabilities)
val = np.arange(0, n)
mode = [int(x) for x in val if round(probabilities[x], 6) == round(mx, 6)]

print(f"\nМатематическое ожидание M(X) = {M:.3f}")
print(f"Дисперсия D(X) = {D:.3f}")
print(f"Среднее квадратическое отклонение b(X) = {b:.3f}")
print(f"Мода Mo(X) = {mode}\n")


plt.figure(figsize=(10, 6))
plt.plot(x, probabilities, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Количество пробирок')
plt.ylabel('Вероятность')
plt.xticks(x)
plt.grid(True, alpha=0.3)
plt.axvline(x=M, color='m', linestyle='--', label=f'Мат. ожидание = {M:.2f}')
plt.axvline(x=mode[0], color='r', linestyle=':', label=f'Мода = {mode}')
plt.legend()
plt.show()
```

Б - закон распределения Пуассона

Предположим, что в воздухе классной комнаты содержится
10^4
бактерий, которые равномерно распределены по всему объёму.
Вероятность нахождения бактерии в случайным образом отобранной пробе
равна в среднем р=10^-3


```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial, exp

def poisson_prob(lmbda, m):
    return (lmbda ** m) * exp(-lmbda) / factorial(m)

n = 10000
p = 0.001
lmbda = n * p

probabilities = []
max_m = 20

for m in range(max_m + 1):
    prob = poisson_prob(lmbda, m)
    probabilities.append(prob)

print("Закон распределения:")
for i in range(max_m + 1):
    print(f"{i:<10} {probabilities[i]:<10.4}")

M = lmbda
D = lmbda
b = np.sqrt(D)
mx = max(probabilities)
val = np.arange(0, max_m)
mode = [int(x) for x in val if round(probabilities[x], 6) == round(mx, 6)]

print(f"\nl = n*p = {lmbda:.3f}")
print(f"Математическое ожидание M(X) = {M:.3f}")
print(f"Дисперсия D(X) = {D:.3f}")
print(f"Среднее квадратическое отклонение b(X) = {b:.3f}")
print(f"Мода Mo(X) = {mode}\n")

x = list(range(max_m + 1))
plt.figure(figsize=(10, 6))
plt.plot(x, probabilities, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Количество бактерий в пробе')
plt.ylabel('Вероятность')
plt.xticks(x)
plt.grid(True, alpha=0.3)
plt.axvline(x=M, color='m', linestyle='--', label=f'Мат. ожидание = {M:.2f}')
plt.axvline(x=mode[0], color='r', linestyle=':', label=f'Мода = {mode}')
plt.legend()
plt.show()
```

В - геометический закон распределения

Программист ищет баг в коде. Каждая проверка участка кода — независимое испытание. Вероятность найти баг на одном участке p = 0.1.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial, exp

def geometric_prob(p, m):
    return p * (1 - p) ** (m - 1)

p = 0.1
max_m = 30

probabilities = []


for m in range(1, max_m + 1):
    prob = geometric_prob(p, m)
    probabilities.append(prob)

print("Закон распределения:")
for i in range(20):
    print(f"{i+1:<10} {probabilities[i]:<10.4}")

M = 1 / p
D = (1 - p) / (p ** 2)
b = np.sqrt(D)
mx = max(probabilities)
val = np.arange(0, max_m)
mode = [int(x) + 1 for x in val if round(probabilities[x], 6) == round(mx, 6)]

print(f"\nl = n*p = {lmbda:.3f}")
print(f"Математическое ожидание M(X) = {M:.3f}")
print(f"Дисперсия D(X) = {D:.3f}")
print(f"Среднее квадратическое отклонение b(X) = {b:.3f}")
print(f"Мода Mo(X) = {mode}\n")

x = list(range(1, max_m + 1))
plt.figure(figsize=(10, 6))
plt.plot(x, probabilities, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Номер попыткии')
plt.ylabel('Вероятность')
plt.xticks(x)
plt.grid(True, alpha=0.3)
plt.axvline(x=M, color='m', linestyle='--', label=f'Мат. ожидание = {M:.2f}')
plt.axvline(x=mode[0], color='r', linestyle=':', label=f'Мода = {mode}')
plt.legend()
plt.show()
```

Г - гипергеометрический закон распределения

В проекте 50 файлов. 10 файлов содержат проблемы с безопасностью. Руководитель проверяет 8 случайных файлов.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

def hypergeom_prob(N, M, n, m):
    return comb(M, m) * comb(N - M, n - m) / comb(N, n)

N = 50
M = 10
n = 8

probabilities = []

for m in range(min(n, M) + 1):
    prob = hypergeom_prob(N, M, n, m)
    probabilities.append(prob)
print("Закон распределения:")
for i in range(9):
    print(f"{i:<10} {probabilities[i]:<10.4}")


Mo = n * M / N
D = (n * M / (N - 1)) * (1 - M / N) * (1 - n / N)
b = np.sqrt(D)
mx = max(probabilities)
val = np.arange(0, int(min(n, M) + 1))
mode = [int(x) for x in val if round(probabilities[x], 6) == round(mx, 6)]

print(f"Математическое ожидание M(X) = {M:.3f}")
print(f"Дисперсия D(X) = {D:.3f}")
print(f"Среднее квадратическое отклонение b(X) = {b:.3f}")
print(f"Мода Mo(X) = {mode}\n")


x = list(range(0, int(min(n, M) + 1)))
plt.figure(figsize=(10, 6))
plt.plot(x, probabilities, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Номер попыткии')
plt.ylabel('Вероятность')
plt.xticks(x)
plt.grid(True, alpha=0.3)
plt.axvline(x=Mo, color='m', linestyle='--', label=f'Мат. ожидание = {M:.2f}')
plt.axvline(x=mode[0], color='r', linestyle=':', label=f'Мода = {mode}')
plt.legend()
plt.show()
```
