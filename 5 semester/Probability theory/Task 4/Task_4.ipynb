```python
from random import randint
import matplotlib.pyplot as plt

def simulate_boy_girl_paradox(num_trials):
    count_at_least_one_boy = 0
    count_both_boys_at_least_one = 0

    count_older_is_boy = 0
    count_both_boys_older_boy = 0

    for _ in range(num_trials):
        choices = [['M', 'F'], ['M', 'M'], ['F', 'M'], ['F', 'F']]
        children = choices[randint(0, 3)]


        # хотя бы один мальчик
        if 'M' in children:
            count_at_least_one_boy += 1
            if children == ['M', 'M']:
                count_both_boys_at_least_one += 1

        # старший - мальчик
        if children[0] == 'M':
            count_older_is_boy += 1
            if children == ['M', 'M']:
                count_both_boys_older_boy += 1

    prob1 = count_both_boys_at_least_one / count_at_least_one_boy
    prob2 = count_both_boys_older_boy / count_older_is_boy

    return prob1, prob2


def visualize(num_trials=1000):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    probabilities = [0.25, 0.25, 0.25, 0.25]

    colors1 = ['green', 'blue', 'blue', 'gray']
    axes[0].bar(['ММ', 'МД', 'ДМ', 'ДД'], probabilities, color=colors1)
    axes[0].set_title('Хотя бы один мальчик')
    axes[0].set_ylabel('Вероятность')
    axes[0].set_ylim(0, 0.3)

    colors2 = ['green', 'blue', 'gray', 'gray']
    axes[1].bar(['ММ', 'МД', 'ДМ', 'ДД'], probabilities, color=colors2)
    axes[1].set_title('Старший - мальчик')
    axes[1].set_ylabel('Вероятность')
    axes[1].set_ylim(0, 0.3)

    plt.tight_layout()
    plt.show()



if __name__ == "__main__":
    print("\nПарадокс двух детей\n")
    prob1, prob2 = simulate_boy_girl_paradox(20000)
    print(f"Если известно, что хотя бы один - мальчик: P(оба мальчики) = {prob1}")
    print(f"Если известно, что старший - мальчик:      P(оба мальчики) = {prob2}")
    print(f"Теоретически 1/3 и 1/2\n")

    visualize()
```
